```python
import pandas as pd

# Load your dataset
df = pd.read_csv("booksDataset_cleaned.csv")

# Remove rows with null values in both "description" and "title" columns
df.dropna(subset=['description', 'title'], how='any', inplace=True)

# Remove leading and trailing whitespace from title and description columns
df['title'] = df['title'].str.strip()
df['description'] = df['description'].str.strip()

# Replace commas with spaces in the "title" column
df['title'] = df['title'].str.replace(',', ' ', regex=False)

# Reset the index after dropping rows
df.reset_index(drop=True, inplace=True)

# Save the cleaned dataset to a new CSV file
df.to_csv("booksDataset_cleaned.csv", index=False)
```
